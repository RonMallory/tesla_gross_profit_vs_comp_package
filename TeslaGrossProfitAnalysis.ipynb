```python
import matplotlib.pyplot as plt
import numpy as np

# Data for years
years = list(range(2011, 2024))

# Financial data for Tesla
gross_profits = [
    62,
    30,
    456,
    882,
    924,
    1599,
    2223,
    4042,
    4069,
    6630,
    13606,
    20853,
    17660,
]
revenues = [
    204,
    413,
    2013,
    3198,
    4046,
    7000,
    11759,
    21461,
    24578,
    31536,
    53823,
    81462,
    96773,
]
operating_incomes = [
    -251,
    -394,
    -61,
    -187,
    -717,
    -667,
    -1632,
    -388,
    -69,
    1994,
    6523,
    13656,
    8891,
]

# Calculate the cumulative profits
cumulative_profits = np.cumsum(gross_profits)

# Create the plot
plt.figure(figsize=(10, 5))
plt.plot(years, gross_profits, marker="o", label="Tesla Gross Profit (in millions)")
plt.plot(
    years,
    cumulative_profits,
    marker="o",
    color="orange",
    label="Cumulative Profit (in millions)",
)
plt.plot(years, revenues, marker="s", color="blue", label="Tesla Revenue (in millions)")
plt.plot(
    years,
    operating_incomes,
    marker="x",
    color="green",
    label="Tesla Operating Income (in millions)",
)
plt.axhline(
    y=56000,
    color="r",
    linestyle="--",
    label="Elon Musk Compensation Request ($56 billion)",
)

# Add labels, title, and grid
plt.title("Tesla Financials vs Elon Musk Compensation Request")
plt.xlabel("Year")
plt.ylabel("Amount (in millions USD)")
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

```python
# percent of total revenue to 56 billion
percent_of_revenue = 56000 / revenues[-1] * 100

# pie chart
plt.figure(figsize=(10, 5))
plt.pie(
    [percent_of_revenue, 100 - percent_of_revenue],
    labels=[
        f"Elon Musk Compensation Request ({percent_of_revenue:.2f}%)",
        "Remaining Revenue",
    ],
    autopct="%1.1f%%",
)
plt.title("Elon Musk Compensation Request as a Percentage of Tesla's Total Revenue")
plt.show()
```

```python
# pie chart percent of total profit to 56 billion
percent_of_profit = 56000 / cumulative_profits[-1] * 100

plt.figure(figsize=(10, 5))
plt.pie(
    [percent_of_profit, 100 - percent_of_profit],
    labels=[
        f"Elon Musk Compensation Request ({percent_of_profit:.2f}%)",
        "Remaining Profit",
    ],
    autopct="%1.1f%%",
)
```
