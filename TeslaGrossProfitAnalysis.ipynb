```python
import matplotlib.pyplot as plt
import numpy as np

# Data for Tesla's gross profits by year
years = list(range(2011, 2024))
gross_profits = [
    62,  # 2011
    30,  # 2012
    456,  # 2013
    882,  # 2014
    924,  # 2015
    1599,  # 2016
    2223,  # 2017
    4042,  # 2018
    4069,  # 2019
    6630,  # 2020
    13606,  # 2021
    20853,  # 2022
    17660,  # 2023
]

# Calculate the cumulative profits
cumulative_profits = np.cumsum(gross_profits)

# Create the plot
plt.figure(figsize=(10, 5))
plt.plot(years, gross_profits, marker="o", label="Tesla Gross Profit (in millions)")
plt.plot(
    years,
    cumulative_profits,
    marker="o",
    color="orange",
    label="Cumulative Profit (in millions)",
)
plt.axhline(
    y=56000,
    color="r",
    linestyle="--",
    label="Elon Musk Compensation Request ($56 billion)",
)

# Add labels, title, and grid
plt.title("Tesla Gross Profit vs Elon Musk Compensation Request")
plt.xlabel("Year")
plt.ylabel("Amount (in millions USD)")
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
